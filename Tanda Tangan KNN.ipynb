# Proses GLCM

```python
def calc_glcm_sudut90(img, label, props=['contrast', 'homogeneity', 'energy', 'correlation'], dists=[1], lvl=256, sym=True, norm=True):
    agls = [np.pi/2]
    glcm = greycomatrix(img,
                        distances=dists,
                        angles=agls,
                        levels=lvl,
                        symmetric=sym,
                        normed=norm)
    feature = []
    glcm_props = [prop for name in props for prop in greycoprops(glcm, name)[0]]
    for item in glcm_props:
        feature.append(item)
    feature.append(label)

    return feature
```

```python
properties = ['contrast', 'homogeneity', 'energy', 'correlation']

glcm_all_agls = []
for img, label in zip(imgs, label):
    glcm_all_agls.append(
            calc_glcm_sudut90(img,
                              label,
                              props=properties)
                            )

columns = []
angles = ['90']
for name in properties :
    for ang in angles:
        columns.append(name + "_" + ang)

columns.append("label")
```

```python
# Membuat DataFrame dari data GLCM
df = pd.DataFrame(glcm_all_agls, columns=columns)
df
```

# Pembagian Data

```python
x = dff.drop('label', axis=1)
y = dff['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

# Euclidean Distance

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
```

```python
knn_euclidean = KNeighborsClassifier(n_neighbors=9, metric='euclidean')
```

```python
knn_euclidean.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# Menghitung prediksi model pada data train
y_pred = knn_euclidean.predict(x_train)

# Menghitung confusion matrix
cm = confusion_matrix(y_train, y_pred)
print("Confusion Matrix:")
print(cm)

# Menghitung akurasi
accuracy = accuracy_score(y_train, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))

# Menghitung presisi
precision = precision_score(y_train, y_pred, average='macro')
print("Precision: {:.2f}%".format(precision * 100))

# Menghitung recall
recall = recall_score(y_train, y_pred, average='macro')
print("Recall: {:.2f}%".format(recall * 100))
```
